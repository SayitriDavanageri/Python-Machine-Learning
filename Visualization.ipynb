#**Numpy**

#create arrays

```python
import numpy as np
a=np.array([1,2,3,4,5])
b=np.arange(1,6) #creates[1,2,3,4,5]
c=np.zeros((2,3)) #2x3 matrix of zeros
d=np.ones((3,2)) #3x2 matrix of ones

print("Array a:",a)
print("Array b:",b)
print("2x3 zeros:",c)
print("3x2 ones",d)
```

#indexing and Slicing

```python
print("First element of a:",a[0])
print("Last two elements of a:",a[-2:])
```

#Basic arithmatic operations

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

#basic statistics

```python
print("square of a:",np.square(a))
print(f"mean of a:{np.mean(a)}")
print(f"standard deviation:{np.std(a)}")
```

#Reshaping arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix:")
matrix
```

#Transpose

```python
print("Transpose:")
matrix.T
```

#boolean indexing

```python
print(f"elements>7:{matrix[matrix>7]}")
```

#**Pandas**

#Create a DataFrame manually

```python
import pandas as pd
data={
    'Name':['Alice','Bob','Reva','Ayla','Jannet'],
    'Age':[20,35,45,27,23],
    'Department':['Sales','Medical','Law','Fine art','Sales'],
    'Salary':[20000,45000,30000,47000,27000]
}
df=pd.DataFrame(data)
df
```

#basic opertaions

```python
print("first 3 rows:")
df.head(3)
```

```python
print("columns names:")
df.columns
```

```python
print("data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

#Filtering

```python
print("people with salary >30000")
df[df['Salary']>30000]
```

#Sorting

```python
print("sorted by age:")
df.sort_values(by='Age')
```

#**Grouping and Aggregation**

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Average salary by Department:")
grouped
```

#Adding new columns

```python
df['Salary in lacks']=df['Salary']/100000
df
```

#**MatPlotLib:Building Blocks of Visualization**

#Line Plot

```python
import matplotlib.pyplot as plt
years=[2015,2016,2017,2018]
revenue=[150,170,195,200]
plt.plot(years,revenue,marker='o',linestyle='-',color='green')
plt.title(" company revenue over years")
plt.xlabel("years")
plt.ylabel("revenue(in crores)")
plt.xticks(years)
plt.grid(False)
plt.show()
```

#Bar chart

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='blue')
plt.title("Employee count by department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

#**Histogram**

```python
ages=df['Age']
plt.hist(ages,bins=5,color='pink',edgecolor='green')
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

#Pie chart

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department proportion")
plt.ylabel("")
plt.show()
```

#**Seaborn:Statistical Data Visualization**

```python
import seaborn as sns
tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\Info about the tips dataset:")
tips.info()
print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

```python
sns.pairplot(tips,hue="day")
plt.show()
```

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("Count of Records by Day")
plt.show()
```

#**Boxplot**

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total bill by day")
plt.show()
```
